```python
import pandas as pd
import numpy as np

from datetime import datetime, timedelta
```

### Create the Dimension for Mobility

```python
cases = pd.read_csv("2020_CA_Region_Mobility_Report.csv")
cases = cases[(cases['date'] >= '2020-07-01') & (cases['date'] < '2020-11-01')]
cases = cases[(cases['sub_region_2'].isin(['Regional Municipality of Peel', 'Regional Municipality of Halton', 'Regional Municipality of York', 'Regional Municipality of Durham', 'Toronto Division', 'Ottawa Division']))]
cases.head(10)
```

```python
cases.columns
```

```python
Mobility_dimension = cases[['date', 'sub_region_1', 'sub_region_2','retail_and_recreation_percent_change_from_baseline', 'grocery_and_pharmacy_percent_change_from_baseline', 'parks_percent_change_from_baseline', 'transit_stations_percent_change_from_baseline', 'workplaces_percent_change_from_baseline', 'residential_percent_change_from_baseline']].copy()
Mobility_dimension.columns = ['Accurate_Episode_Date', 'Province', 'Subregion', 'Retail_And_Recreation', 'Grocery_And_Pharmacy', 'Parks', 'Transit_Stations', 'Workplaces', 'Residential']
Mobility_dimension.head(1000)
```

```python
Mobility_dimension.insert(0, "Mobility_surrogate_key",  np.arange(len(Mobility_dimension)))
Mobility_dimension.head(70)
```

```python
Mobility_dimension['Parks'].fillna(Mobility_dimension['Parks'].mode()[0], inplace=True)
Mobility_dimension['Transit_Stations'].fillna(Mobility_dimension['Transit_Stations'].mode()[0], inplace=True)
Mobility_dimension.head(70)
```

```python
Mobility_dimension.to_csv('Mobility_dimension_with_dates.csv',index = False, header=True)
```
