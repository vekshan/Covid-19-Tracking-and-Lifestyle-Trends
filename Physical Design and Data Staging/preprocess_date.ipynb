```python
import pandas as pd
import numpy as np
import holidays

from datetime import datetime, timedelta
```

```python
canada_ontario_holiday_dates_2020_2021 = {}

for date in holidays.Canada(years=[2020, 2021]).items():
    canada_ontario_holiday_dates_2020_2021[str(date[0])] = str(date[1])

seasons = {1:"Winter", 2:"Spring", 3:"Summer", 4:"Autumn"}

df_mar_2020_feb_2021 = pd.DataFrame({"Full_Date": pd.date_range(start="2020-03-02", end="2021-02-28")})
df_mar_2020_feb_2021["Week_Day_Name"] = df_mar_2020_feb_2021.Full_Date.dt.day_name()
df_mar_2020_feb_2021["Day"] = df_mar_2020_feb_2021.Full_Date.dt.day
df_mar_2020_feb_2021["Month"] = df_mar_2020_feb_2021.Full_Date.dt.month
df_mar_2020_feb_2021["Year"] = df_mar_2020_feb_2021.Full_Date.dt.year
df_mar_2020_feb_2021["Day_Of_Year"] = df_mar_2020_feb_2021.Full_Date.dt.dayofyear
df_mar_2020_feb_2021["Day_Of_Week"] = df_mar_2020_feb_2021.Full_Date.dt.dayofweek
df_mar_2020_feb_2021["Week_In_Year"] = df_mar_2020_feb_2021.Full_Date.dt.isocalendar().week
df_mar_2020_feb_2021["Quarter"] = df_mar_2020_feb_2021.Full_Date.dt.quarter
df_mar_2020_feb_2021["Is_Weekend"] = df_mar_2020_feb_2021.Day_Of_Week.map(lambda day: 1 if day >= 5 else 0)
df_mar_2020_feb_2021["Is_Week_Day"] = df_mar_2020_feb_2021.Day_Of_Week.map(lambda day: 1 if day < 5 else 0)
df_mar_2020_feb_2021["Is_Holiday"] = df_mar_2020_feb_2021.Full_Date.map(lambda date: 1 if str(date).split()[0] in canada_ontario_holiday_dates_2020_2021 else 0)
df_mar_2020_feb_2021["Holiday_Name"] = df_mar_2020_feb_2021.Full_Date.map(lambda date: canada_ontario_holiday_dates_2020_2021[str(date).split()[0]] if str(date).split()[0] in canada_ontario_holiday_dates_2020_2021 else None)
df_mar_2020_feb_2021["Season"] = df_mar_2020_feb_2021.Month.map(lambda month: seasons[(month%12 + 3)//3])

df_mar_2020_feb_2021.insert(0,  "Date_surrogate_key", np.arange(100, len(df_mar_2020_feb_2021) + 100))
                            
df_mar_2020_feb_2021.head(100)
```

```python
df_mar_2020_feb_2021.to_csv("Date_dimension.csv", index = False, header = True)
```
