```python
import pandas as pd
import numpy as np

from datetime import datetime, timedelta
```

### Create the dimension for Weather Ottawa

```python
cases = pd.read_csv("en_climate_daily_ON_6105976_2020_P1D.csv")
cases = cases[(cases['Date/Time'] >= '2020-07-01') & (cases['Date/Time'] < '2020-11-01')]
cases.head(10)
```

```python
cases.columns
```

```python
Weather_dimension_Ottawa = cases[['Station Name', 'Date/Time', 'Max Temp (°C)','Min Temp (°C)', 'Total Precip (mm)']].copy()
Weather_dimension_Ottawa.columns = ['Station_Name', 'Accurate_Episode_Date', 'Daily_High_Temperature', 'Daily_Low_Temperature', 'Precipitation']
Weather_dimension_Ottawa.head(100)
```

### Create the dimension for Weather Toronto

```python
cases = pd.read_csv("en_climate_daily_ON_6158731_2020_P1D.csv")
cases = cases[(cases['Date/Time'] >= '2020-07-01') & (cases['Date/Time'] < '2020-11-01')]
cases.head(10)
```

```python
cases.columns
```

```python
Weather_dimension_Toronto = cases[['Station Name', 'Date/Time', 'Max Temp (°C)','Min Temp (°C)', 'Total Precip (mm)']].copy()
Weather_dimension_Toronto.columns = ['Station_Name', 'Accurate_Episode_Date', 'Daily_High_Temperature', 'Daily_Low_Temperature', 'Precipitation']
Weather_dimension_Toronto.head()
```

### Create the dimension for Weather

```python
Weather_dimension = pd.concat([Weather_dimension_Toronto, Weather_dimension_Ottawa])
Weather_dimension.head()
```

```python
Weather_dimension.insert(0, "Weather_surrogate_key",  np.arange(len(Weather_dimension)))
Weather_dimension.head(100)
```

```python
Weather_dimension.to_csv('Weather_dimension_with_dates.csv',index = False, header=True)
```
