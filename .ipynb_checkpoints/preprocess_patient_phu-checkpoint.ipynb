```python
import pandas as pd
import numpy as np

from datetime import datetime, timedelta
```

### Create the dimension for Weather Ottawa

```python
cases = pd.read_csv("en_climate_daily_ON_6105976_2020_P1D.csv")
cases = cases[(cases['Date/Time'] >= '2020-07-01') & (cases['Date/Time'] < '2020-11-01')]
cases.head(10)
```

```python
cases.columns
```

```python
Weather_dimension_Ottawa = cases[['Station Name', 'Date/Time', 'Max Temp (°C)','Min Temp (°C)', 'Total Precip (mm)']].copy()
Weather_dimension_Ottawa.columns = ['Station_Name', 'Accurate_Episode_Date', 'Daily_High_Temperature', 'Daily_Low_Temperature', 'Precipitation']
Weather_dimension_Ottawa.head(100)
```

```python
Weather_dimension_Ottawa.insert(0, "Weather_Surrogate_Key",  np.arange(len(Weather_dimension_Ottawa)))
Weather_dimension_Ottawa.head(100)
```

```python
Weather_dimension_Ottawa.to_csv('test.csv',index = False, header=True)
```

### Create the Dimension for Mobility

```python
cases = pd.read_csv("2020_CA_Region_Mobility_Report.csv")
cases = cases[(cases['date'] >= '2020-07-01') & (cases['date'] < '2020-11-01')]
cases = cases[(cases['sub_region_2'].isin(['Ottawa Division', 'Toronto Division']))]
cases.head(10)
```

```python
cases.columns
```

```python
Mobility_dimension = cases[['date', 'metro_area', 'sub_region_1', 'sub_region_2','retail_and_recreation_percent_change_from_baseline', 'grocery_and_pharmacy_percent_change_from_baseline', 'parks_percent_change_from_baseline', 'transit_stations_percent_change_from_baseline', 'workplaces_percent_change_from_baseline', 'residential_percent_change_from_baseline']].copy()
Mobility_dimension.columns = ['Accurate_Episode_Date', 'Metro_Area', 'Province', 'Subregion', 'Retail_And_Recreation', 'Grocery_And_Pharmacy', 'Parks', 'Transit_Stations', 'Workplaces', 'Residential']
Mobility_dimension.head(1000)
```

```python
Mobility_dimension.insert(0, "Mobility_Surrogate_Key",  np.arange(len(Mobility_dimension)))
Mobility_dimension.head(100)
```

```python
Mobility_dimension.to_csv('test3.csv',index = False, header=True)
```
